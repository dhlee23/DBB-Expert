### changes in the Return to Education and the Gender Wage Gap

```python
import wooldridge as woo
import pandas as pd
import statsmodels.formula.api as smf
```

```python
# current population survey

cps78_85 = woo.dataWoo('cps78_85')

cps78_85.head()
```

```python
cps78_85.describe()
```

```python
# independent variable : log hourly wage

reg = smf.ols(formula = 'lwage ~ y85*(educ + female) + exper + expersq + union',
             data = cps78_85)

results = reg.fit()

results.summary()

# dummy variable들끼리의 결합은 모두 절편을 이룬다
```
