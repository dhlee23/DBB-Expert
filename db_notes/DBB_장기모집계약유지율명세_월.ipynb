```python
# 증권번호로 그룹화, 기준년월 기준으로 정렬
df_1 = df.sort_values(by=["증권번호","기준년월"])[col1]

# 직전 보유계약건수 값을 확인하기 위해 shift() 함수 사용
df_1["previous_s"] = df_1.groupby("증권번호")['보유계약건수'].shift(1)

# 보유계약건수가 1에서 0으로 바뀌는 패턴 찾기
df_2 = df_1[(df_1['보유계약건수']==0) & (df_1['previous_s']==1)]

print(len(set(df_2['증권번호'])))
print(len(df_2))
```
