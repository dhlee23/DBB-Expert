## 기본적인 시각화 객체 생성

```python
# figure와 subplot : 같은 셀에서 실행해야 함

# 1. figure 객체를 만듦 : 크기 지정 가능

fig = plt.figure()

# 2. 하나 이상의 subplot 생성

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax3.plot(np.random.standard_normal(50).cumsum(), color = 'black', linestyle  = 'dashed')
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax1.plot(np.random.standard_normal(50).cumsum(), color = 'black', linestyle  = 'dashed')

ax2.hist(np.random.standard_normal(100), bins = 20, color='blue', alpha=0.5)

ax3.scatter(np.arange(30), np.arange(30) + 3*np.random.standard_normal(30))
```

```python
# subplot을 2 x 3 으로 정의
# 개별적으로 만들지 않았으므로, axes[0, 0], axes[0,1] 등으로 호출

fig, axes = plt.subplots(2,3) 

axes[0,1].hist(np.random.standard_normal(100), bins = 20, color='blue', alpha=0.5)
```

```python
# 하나만 그릴 때도 subplot을 만드는 편이 낫다

fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(np.arange(30), np.arange(30) + 3*np.random.standard_normal(30))
```

```python
# subplot간의 간격 조절

fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax3.plot(np.random.standard_normal(50).cumsum(), color = 'black', linestyle  = 'dashed')

ax1.hist(np.random.standard_normal(100), bins = 20, color='blue', alpha=0.5)

ax2.scatter(np.arange(30), np.arange(30) + 3*np.random.standard_normal(30))

# subplots_adjust

plt.subplots_adjust(left=1, bottom=None, right=3, top=None, wspace=0.2, hspace=1)

# wspace와 hspace는 서르플롯간의 간격을 위해 너비와 높이의 비율을 조정
```

```python
# 서브프롯간의 간격을 주지 않는 그래프를 생성

fig, axes = plt.subplots(2,2, sharex=True, sharey=True)

for i in range(2):
    for j  in range(2):
        axes[i,j].hist(np.random.standard_normal(500), bins=50, color='blue', alpha=0.5)

fig.subplots_adjust(wspace=0, hspace=0)
```

```python
# 한글폰트, color, linesyle, marker

plt.rcParams['font.family'] ='AppleGothic'
plt.rcParams['axes.unicode_minus'] =False

fig = plt.figure()

ax = fig.add_subplot()

ax.set_title("시험 플롯")

ax.plot(np.random.standard_normal(30).cumsum(), color = 'red', linestyle="--", marker='o')

plt.show()
```
