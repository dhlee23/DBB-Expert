```python
import pandas as pd

df = pd.read_csv('./weather_tmp.csv', encoding = 'utf-8')
df.head()
```

## 인덱스와 칼럼의 변경 (칼럼을 인덱스로, 인덱스를 칼럼으로)

```python
# 만약 우리가 현재 0부터 시작하는 rangeIndex대신에
# '일시' 칼럼을 인덱스로 하기를 원할 수도 있다. 그럴 때...
df1 = df.copy()
df1.set_index('일시', inplace = True)
```

```python
df1.head()
```

```python
df.head()
```

```python
df.set_index('일시')
```

```python
# 만약 인자값에 drop = False 를 넣게 된다면, 원래의 '일시'가 인덱스로 전환됨과 함께 칼럼에도 남아 있게 된다.
# 기본값은 drop = True이다.

df.set_index('일시', drop = False)
```

```python
#여기에 인자로 inplace = True를 넣게 되면 원본 df가 변하게 된다

df.set_index('일시', inplace= True)
df.head()
```

```python
# 다시 원래의 0부터 시작하는 인덱스로 돌려 놓아 보자.
# 현재는 그 값들이 따로 저장되어 있지 않으나, 원래의 기본 인덱스이므로 간단히...

df.reset_index(inplace=True)  #inplace = True로 원래의 값으로 완전히 돌려 놓았다
df.head()
```

```python
df1 = df.set_index('일시')
df1.head()
```

## 인덱스와 칼럼의 정렬

```python
# 여기서 인덱스인 일시를 역순으로 돌려보자. 12월 31일이 처음에 오게...

df1.sort_index(ascending=False)
```

```python
#만약 칼럼명도 가나다 순으로 바꾸기를 원한다면..
# 칼럼명도 넓은 의미에서는 인덱스와 속한다고 할 수 있으므로,,,,
# column축을 의미하는 axis = 1 을 명시하면 된다

df1.sort_index(ascending=True, axis = 1)
```

## 값에 따른 정렬

```python
# 만약 '최고기온'을 기준으로 가장 더웠던 날 5일의 날씨를 보려면 어떻게 할까
# '최고기온' 칼럼을 값을 기준으로 다음과 같이 정렬하면 된다

temp = df1.sort_values(by ='최고기온', ascending = False)
temp.head()
```

```python
# 최고기온이 가장 낮은 5일은

temp.tail(5)
```

## 열과 행의 삭제

```python
#만약 '합계일조시간' 칼럼 전체를 삭제하기를 원하면
# 칼럼이라는 것을 명시하는 axis =1을 넣어

temp.drop('합계일조시간', axis=1)
```

```python
# 만약 여러개 칼럼을 동시에 삭제하기를 원하면 []안에 같이 명시하면 된다.
# 주의할 것은 inplace = True를 명시하지 않는 한, 원본 데이터프레임은 변하지 않는다는 것

temp.drop(['최고기온','최저기온'], axis = 1)
```

```python
# 만약 행(row)를 삭제하기를 원한다면, 똑같다.

df.head()
```

```python
#삭제할 행의 인덱스를 넣어주어서...
# 이때 axis = 0 을 넣어도 되고 안 넣어도 된다. default가 axis = 0 이므로.

df.drop([0,2,4])
```

```python
# 다만 drop의 경우, 슬라이싱에 의한 지정이 안 되므로, 개별적으로 하나하나 값을 
# 리스트 형식으로 적어주어야 하므로 번거로운 면이 없지 않다. 
# 이 점에서 나중에 배우게 될 조건 지정 추출 등이 중요하게 된다.
```

```python
#기타 하나의 칼럼만을 삭제할 경우 아래와 같은 del '삭제할 칼럼'으로 하기도 하고..

del df['지점']

df.head()
```

```python
# 삭제하는 동시에 삭제되는 칼럼을 반환하는  df.pop('칼럼명')을 쓰는 경우도 있지만..
# 자주 사용되지는 않는다.

df.pop('일시')
```
