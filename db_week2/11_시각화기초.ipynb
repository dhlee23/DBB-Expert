```python
import pandas as pd

df = pd.read_csv('./weather_final.csv', encoding='utf-8')
print(len(df))
df.head()
```

```python
#넘파이와 맷플롯립을 임포트하자

import matplotlib.pyplot as plt
import numpy as np

import matplotlib 

%matplotlib inline

plt.rcParams["figure.figsize"] = (10,6)
```

## 기본 사용법

```python
#간단한 plot를 그려보자

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# 선이 아닌 marker(여기서는 'o'으로 표시)

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
# x 값을 부여하지 않을 경우, 자동적으로 0부터 y의 갯수로 해석한다

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

```python
#마커와 라인을 결합하면 다음과 같다

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker='*')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o--b')
plt.show()
```

## line plot

```python
seoul = df[df['지점'] == 108]['평균기온']
ypoints = seoul
xpoints = df.iloc[:365]['일시']
xlabels = xpoints
plt.plot(xpoints,ypoints)

plt.xticks(ticks=xpoints, labels=xlabels, rotation=45)
plt.xlabel('2021 Daily Trend')
plt.locator_params(axis='x', nbins=len(xlabels)/30)
plt.show()
```

```python
# 두선을 겹치고 label(lengend)을 표시해 보자
# 주의할 점 : busan과 jeju는 현재 데이터프레임이고, index를 가진다
# 만약 인덱스를 그대로 둔 채, 그래프에 입력하며 두 그래프가 분리된다.
# 인덱스를 제거하는 것이 필요. .values, to_numpy(), to_list()

busan = df[df['지점'] == 159]['평균기온']
jeju = df[df['지점'] == 189]['평균기온']


plt.plot(busan.to_numpy(), label='Seoul')
plt.plot(jeju.to_numpy(), label ='Jeju')
plt.legend()
plt.show()
```

```python
# x축과 y축의 이름을 표시해 보자

plt.plot(busan.to_numpy(), label='Busan')
plt.plot(jeju.to_list(), label ='Jeju')

plt.xlabel("date")
plt.ylabel("Avg Temperature")

plt.legend()
plt.show()
```

```python
# 그래프의 타이틀을 표시해 보자

plt.plot(busan.to_numpy(), label='Busan')
plt.plot(jeju.to_list(), label ='Jeju')

plt.title("Temperature- Busan vs Jeju")
plt.xlabel("date")
plt.ylabel("Avg Temperature")

plt.legend()
plt.show()
```

## scatter plot

```python
#만약 선이 아니라, 점으로 표시하자면 plot()을 scatter()로 표시하고....
# plot()과 달리, scatter()는 반드시 x 값을 요하므로 지정해 주어야 한다.

x = range(1,366)

plt.scatter(x, busan.to_numpy(), label='Busan')
plt.scatter(x, jeju.to_list(), label ='Jeju', alpha = 0.4) #투명도도 조정하였다

plt.title("Temperature- Busan vs Jeju")
plt.xlabel("date")
plt.ylabel("Avg Temperature")

plt.legend()
plt.show()
```

```python
# subplot을 만드는 방식
# 동일한 그림을 그리는 두 개의 방법

x = range(1,366)

fig, ax = plt.subplots()

ax.scatter(x, busan.to_numpy(), label='Busan')
ax.scatter(x, jeju.to_list(), label ='Jeju', alpha = 0.4) #투명도도 조정하였다

ax.set_title("Temperature- Busan vs Jeju")
ax.set_xlabel("date")
ax.set_ylabel("Avg Temperature")

ax.legend()
plt.show()
```

## histogram

```python
# 히스토그램의 가장 큰 특징은 y값을 필요로 하지 않고, 한 변수의 빈도만을 자동 계산하여 표시한다는 것이다

# np.random.randn 로 정규분규를 random하게 생성해 보자
x = np.random.randn(10000)
plt.hist(x, bins= 20)
plt.show()
```

```python
plt.hist(seoul,alpha=0.5, label = 'seoul', bins=50)
plt.legend()
plt.show() 
```

```python

plt.hist(seoul,alpha=.5, label = 'seoul', bins= 25)
plt.hist(jeju,alpha=.5, label = 'jeju', bins = 25)
plt.legend()
plt.show() 
```

```python
fig, axes = plt.subplots(1,2)

axes[0].hist(seoul,alpha=.5, label = 'seoul', bins= 25, color='red')
axes[1].hist(jeju,alpha=.5, label = 'jeju', bins = 25)
axes[0].legend()
axes[1].legend()
plt.show()
```

## bar plot

```python


years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(years, values, color = ['r', 'g', 'b'])

plt.show()
```

```python
# horizontal bar plot

years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(years, values)

plt.xlabel("Numbers")
plt.ylabel("Year", rotation = 0)

plt.show()
```
