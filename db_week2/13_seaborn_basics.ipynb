```python
#라이브러리 임포트
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

#seaborn에서 제공하는 flights 데이터 셋을 사용
flights = sns.load_dataset('flights')

#그래프 사이즈 설정
plt.figure(figsize=(12, 4))
```

```python
flights
```

## Box plot

```python
sns.boxplot(data=flights, x="year", y="passengers")
```

## Violin plot

```python
sns.violinplot(data=flights, x="year", y="passengers")
```

## line plot

```python
sns.lineplot(data=flights, x="year", y="passengers")
```

```python
sns.distplot(flights["passengers"], kde = True)
plt.show()
```

## bar plot

```python
sns.barplot(data=flights, x="year", y="passengers")
plt.show()
```

## 여러개 그리기

```python
fig, axes = plt.subplots(2,1)

sns.boxplot(data=flights, x="year", y="passengers", ax = axes[0])
sns.barplot(data=flights, x="year", y="passengers", ax = axes[1])

plt.show()
```
